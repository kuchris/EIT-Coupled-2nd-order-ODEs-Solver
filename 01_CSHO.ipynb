```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
x = np.linspace(-1.5,1.5,200000)

F = 0.01
w = 0.5
b = 1e-7
a = 4e-2
c1=0.5001
```

```python
def f(x,c):
    return (-2*(np.pi)*1j*F*x*(w**2-x**2-1j*b*x))/(((w**2-x**2-1j*a*x)*(w**2-x**2-1j*b*x))-c**4)
```

```python
plt.figure()

plt.title('Absorbed probe power')
plt.plot(x,np.real(f(x,c1)), label='Real part')
plt.plot(x,np.imag(f(x,c1)), label='Imag part')
plt.ylabel('Absorbed probe power P/m  / Dispersion')
plt.xlabel('Detuning')

#plt.xlim(0.25,0.75)

plt.tight_layout()
```
