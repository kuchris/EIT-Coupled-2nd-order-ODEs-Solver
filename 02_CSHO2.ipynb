```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.style.use(['science','nature','notebook'])
```

```python
x = np.linspace(-10,10,1000000)

F = 0.1
w = 2
a = 4.0*10e-2
b = 1.0*10e-7


c1=0.5

# zeta2 = 1e-7
# zeta1 = 4e-2
# omega0 = 3.15
# omegas = 0
# #Omega_r = 0.35
# F=0.01
# Mass=1
```

```python
def f(x,c):
    return (-2*(np.pi)*1j*(F)*x*(w**2-x**2-1j*b*x))/(((w**2-x**2-1j*a*x)*(w**2-x**2-1j*b*x))-c**4)


#def f(x,c):
#    return (-2*(np.pi)*1j*F*w*(x**2-w**2-1j*b*w))/(((x**2-w**2-1j*a*w)*(x**2-w**2-1j*b*w))-c**4)
```
